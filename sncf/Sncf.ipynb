```python
import pandas as pd
stops = pd.read_csv("data_sncf/stops.csv")
stop_times = pd.read_csv("data_sncf/stop_times.csv")
trips = pd.read_csv("data_sncf/trips.csv")
```

```python
stop_complete_info = stops.merge(stop_times,on="stop_id")
```

Tout d'abord on fusionne les tables stops et stop_times afin d'obtenir une table avec des informations plus complète tel quel le trip_id

```python
stop_complete_info.head()
```

```python
stop_trip_complete_info = stop_complete_info.merge(trips,on="trip_id")
```

A partir de là on fusionne notre table stop_trip_complete avec la table des trips afin d'obtenir les informations  complète sur un trip et ses stops

```python
stop_trip_complete_info.head()
```

```python
from_gare_de_l_est_before_ten = stop_trip_complete_info[(stop_trip_complete_info["stop_name"] == "Gare de Paris-Est") & (stop_trip_complete_info["departure_time"] <= "10:00:00")][["trip_id", "departure_time"]]
```

On récupère tout les trips ayant un arrêt à la gare de Paris-Est et dont l'horaire de départ est fixé avant 10:00 du matin

```python
from_gare_de_l_est_before_ten.sort_values("departure_time")
```

On peut donc voir qu'il y a 70 trains qui partent de la gare de l'est avant 10h. Le premier à 05:42 et le dernier à 09:59.
